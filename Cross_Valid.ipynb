```python
%matplotlib inline
from sklearn.datasets import load_boston as boston
from sklearn.linear_model import LinearRegression as lr
from sklearn.model_selection import cross_val_predict as cross;import matplotlib.pyplot as plt
y=boston().target

predicted=cross(lr(),boston().data,y,cv=10)
fig,ax=plt.subplots()
ax.scatter(y,predicted,edgecolors=(0,0,0))
ax.set_xlabel('Messured')
ax.set_ylabel('Predicted')
plt.show(ax.plot([y.min(),y.max()],[y.min(),y.max()],'k--',lw=4))
```
