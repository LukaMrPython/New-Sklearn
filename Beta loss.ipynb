```python
%matplotlib inline
import numpy as np,matplotlib.pyplot as plt
from sklearn.decomposition.nmf import _beta_divergence as beta

x=np.linspace(.001,4,1000)
y=np.zeros(x.shape)

colr='mbgyr'
for j,b in enumerate((0.,.5,1.,1.5,2.)):
    for i,xi in enumerate(x):
        y[i]=beta(1,xi,1,b)
    plt.plot(x,y,label='beta=%1.1f'%b,color=colr[j])
plt.xlabel('x')
plt.title('beta-divergence(1,x)')
plt.legend(loc='best')
plt.axis([0,4,0,3])
```
